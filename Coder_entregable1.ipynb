## Paso 2: Importar libreria de yahooo finance y descargar el historico de tesla

```python
import yfinance as yf
tsla = yf.Ticker('TSLA')
#tsla.info
```

```python
## sacar la informacion historica de un a√±o hacia atras
hist = tsla.history(period = '1y')
hist['Date'] = hist.index
hist = hist.reset_index(drop = True)
```

```python
hist.columns
```

## Paso 3

```python
# extraemos graficos rapidos
tickers = ['TSLA']
import matplotlib.pyplot as plt
for i, ticker in enumerate(tickers):
  current_ticker  = yf.Ticker(ticker)
  plt.subplot(len(tickers), 1, i+1)
  current_ticker.history(period = '365d')['Close'].plot(figsize = (16, 10), title = 'Precio historico de un anio para: '+ticker)
```
